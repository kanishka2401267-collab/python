```python
def count_vowels(s):
    count = 0
    for ch in s.lower():
        if ch in "aeiou":
            count += 1
    return count
print(count_vowels("Hello World"))
```

```python
students = {
    "Anu": [78, 85, 90],
    "Bala": [60, 72, 68],
    "Charan": [88, 91, 84],
    "Divya": [45, 50, 40]
}


def calculate_average(marks):
    return sum(marks) / len(marks)

check_result = lambda avg: "Pass" if avg >= 50 else "Fail"


student_result = {}

for name, marks in students.items():
    avg = calculate_average(marks)
    student_result[name] = {
        "Average": round(avg, 2),
        "Result": check_result(avg)
    }

print("Passed Students:")
for name, details in student_result.items():
    if details["Result"] == "Pass":
        print(name, details)


topper = max(student_result, key=lambda x: student_result[x]["Average"])
print("Topper:", topper)
```

```python
def count_digits(n):
    if n == 0:
        return 0
    return 1 + count_digits(n // 10)
n = 48392
print("Digits:", count_digits(n))

```

```python
def sum_digits(n):
    if n == 0:
        return 0
    return (n % 10) + sum_digits(n // 10)
n = 48392
print("Sum:", sum_digits(n))
```

```python
def max_digit(n):
    if n < 10:
        return n
    return max(n % 10, max_digit(n // 10))
n = 48392
print("Max digit:", max_digit(n))
```

```python
def reverse_number(n, rev=0):
    if n == 0:
        return rev
    return reverse_number(n // 10, rev * 10 + n % 10)
n = 48392
print("Reverse:", reverse_number(n))
```
